```python
from bs4 import BeautifulSoup
```

```python
import requests
import datetime
import pandas as pd
import re

```

```python
url = "https://en.wikipedia.org/wiki/2016%E2%80%932022_Yemen_cholera_outbreak"
```

```python
result=requests.get(url)
doc = BeautifulSoup(result.text,"html.parser")
```

```python
print (result.text)
```

```python
print(len(doc.find_all('table')))
```

```python
table_doc = doc.find_all('table')
```

```python
print(table_doc)
```

```python
# looking for the specific table called wikitable on the wikipedia html
tables=doc.find("table",class_="wikitable")
```

```python
print(tables)
```

```python
# finding all the headers names from table header(th) in wikitable file
headers=tables.find_all('th')
```

```python
headers
```

```python
titles =[]
for i in headers:
    title =i.text
    titles.append(title.strip())
```

```python
print(titles)
```

```python
df=pd.DataFrame(columns=titles)
```

```python
df
```

# The individual values under the headers names

```python
rows=tables.find_all('tr')
```

```python
rows
```

```python
#Finding all the table rows(tr) and table data(td) in the wilkitable file
for i in rows[1:]:
    data=i.find_all("td")
    #print(data)
    row= [tr.text for tr in data]
    #print(row)
    l=len(df)
    df.loc[l]= row
   
    
```

```python
print(df)
```

```python

df.to_csv("yemen_cholera.csv")
    
```
